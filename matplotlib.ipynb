```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#with line
xpoi = np.array([0,5,8])
ypoi = np.array([0,12,32])
plt.plot(xpoi,ypoi)
plt.show()
```

```python
#without line
xpoi = np.array([2,6])
ypoi = np.array([15,5])
plt.plot(xpoi,ypoi,'*')
#o-circle; *-star; .-point ; ,-pixel; x - X; X - X(filled); + - plus; P - plus ; s - square; D - diamond; p - pentagon; H,h,v,^,<,>,1,2,4,3,|,_
plt.show()
```

```python
#ploting multiple points
xpoin = np.array([1,2,7,4,4])
ypoin = np.array([3,8,10,9,5])
plt.plot(xpoin, ypoin, '.')
plt.show()
```

```python
#without explicit points for x-axis
ypoi = np.array([4,6,3,7,9])
plt.plot(ypoi,marker = 'o')
plt.show()
```

```python
#format string fmt
ypoi = np.array([2,6,4,7,9])
plt.plot(ypoi, 'o-.r')#o - circle, - dashed line, r - red
plt.show()
```

```python
#marker style
ypoi = np.array([3,5,2,8])
plt.plot(ypoi, marker = 'D', ms =10 ,mec = 'g',mfc = 'y')#ms - size of circle
plt.show()
```

```python
#line style
ypoi = np.array([3,6,8,9])
plt.plot(ypoi, linestyle = 'dashed', color = 'blue', linewidth='3')
plt.show()
```

```python
#multiple lines
y1 = np.array([3,8,1,10])
y2 = np.array([6,4,7,9])
y3 = np.array([3,7,2,1])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.show()
```

```python
x1 = np.array([0,3,6,8])
x2 = np.array([2,5,4,9])
y1 = np.array([3,6,8,2])
y2 = np.array([2,5,8,4])
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
n1=np.array([20,30,48,57,64,43,32,123,100])
n2=np.array([210,323,125,290,175,142,251,321,198])

plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("Plot title")

plt.plot(n1,n2)
plt.show()
```

```python
#formatted
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("X-Label", fontdict = font1)
plt.ylabel("Y-Label", fontdict = font2)
plt.title("PLot title", fontdict = font2)

plt.plot(x,y)
plt.show()
```

```python
#add grid line
plt.plot(x,y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
