```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
data.head()
```

```python
sns.get_dataset_names()
```

```python
sns.lineplot(x='sepal_length', y='sepal_width',data=data)
```

```python
plt.title('Title using Matplotlib Function')
```

```python
plt.xlim(4,8)
```

```python
#changing theme
sns.lineplot(x='sepal_length',y='sepal_width', data=data)
sns.set_style("ticks")#(darkgrid/whitegrid/dark/white/ticks)
plt.show()
```

```python
#removing spines
sns.lineplot(x='sepal_length', y='sepal_width',data=data)
sns.despine()
plt.show()
```

```python
#changing fig size
plt.figure(figsize =(10,6))#width and height of the plot
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
sns.despine()
plt.show()
```

```python
#scaling
sns.set_context("talk")
plt.show()
```

```python
#current color palatte
palette = sns.color_palette()
sns.palplot(palette)
plt.show()
```

```python
#diverging color palette PiYG,PRGn,BrBG,seismic,RdYIGn,bwr
palette = sns.color_palette('Spectral',15)#blue-white-red - 15 distinict colors
sns.palplot(palette)
plt.show()
```

```python
#sequential color palette Greens,Oranges,Reds,YlOrBr, OrRd,BuGn,..
palette = sns.color_palette('PuBuGn',20)
sns.palplot(palette)
plt.show()
```

```python
#barplot, loading dataset
data = sns.load_dataset("iris")
sns.barplot(x='species',y='sepal_length',data=data)
plt.show()
```

```python
#countplot
data=sns.load_dataset("iris")
sns.countplot(x='species',data=data)
plt.show()
```

```python
#boxplot
sns.boxplot(x='species', y='sepal_width',data=data)
plt.show()
```

```python
#kerneldensityplot
sns.kdeplot(data,fill="True")
plt.show()
```

```python
#violinplot = features of box and kernel density
sns.violinplot(x='species',y='sepal_width',data=data)
plt.show()

```

```python
#stripplot - indi datapts along single axis
data = sns.load_dataset("iris")
sns.stripplot(x='species', y='sepal_width',data=data)
plt.show()
```

```python
#histogram
data=sns.load_dataset("iris")
sns.histplot(x='species',y='sepal_width',data=data)
plt.show()
```

```python
#distplot
sns.distplot(data['sepal_length'])
plt.show()
```

```python
#pairplot - explore the relationships between multiple pairs of variables in a dataset
#hue -elationships between variables while considering the impact of a categorical factor.
sns.pairplot(data=data, hue="species", markers=["o","s","D"],corner=True)
plt.show()
```
